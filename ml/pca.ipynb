```python
from sklearn.datasets import fetch_olivetti_faces
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from sklearn.decomposition import PCA
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = fetch_olivetti_faces()
```

```python
pca = PCA(n_components=150)
```

```python
lower_dimensional_data = pca.fit_transform(data.data)
```

```python
pca.n_components_
```

```python
approximation = pca.inverse_transform(lower_dimensional_data)
```

```python
plt.figure(figsize=(8,4));
x = 276
# Original Image
plt.subplot(1, 2, 1);
plt.imshow(data.images[x])
plt.xlabel('784 components', fontsize = 14)
plt.title('Original Image', fontsize = 20);

# 154 principal components
plt.subplot(1, 2, 2);
plt.imshow(approximation[x].reshape(64,64))
plt.xlabel('123 components', fontsize = 14)
plt.title('95% of Explained Variance', fontsize = 20);
```

```python
sum(pca.explained_variance_ratio_)
```

```python
# approximation
X_train,X_test,Y_train,Y_test = train_test_split(approximation,data.target,test_size=0.3,random_state=100)
```

```python
model = LogisticRegression()
```

```python

model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```
